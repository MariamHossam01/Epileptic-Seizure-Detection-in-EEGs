```python
from scipy.stats import skew
import mne
import numpy as np
import os
```

```python
# Load the EEG data in EDF format
raw = mne.io.read_raw_edf('data/chb01_04.edf', preload=True)
raw.load_data()
raw.filter(0.5, 30)
raw.set_eeg_reference('average')
```

```python
raw.plot()
```

```python
raw.plot_psd(fmin=0, fmax=40)
```
